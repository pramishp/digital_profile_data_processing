```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
pd.options.display.max_columns = None
```

```python
df = pd.read_excel('family_basic_information_-_latest_version_-_English_En_-_2021-03-24-06-35-58.xlsx')
df.columns = [column[:10] for column in df.columns]
df.head()
```

```python
# list(df.columns)
```

```python
# table = pd.pivot_table(data=df,index=['Gender'])
# table

pd.crosstab(index=df['Religion'], columns=df['Gender'])
```

```python
pd.crosstab(index=df['Religion'], columns=df['Ethnicity'], values=df['Ethnicity'], aggfunc=pd.Series.nunique)
```

```python
pd.crosstab(index=df['Mother Ton'], columns=[df['Occupation'], df['Ward Numbe']], values=df["Total numb"], aggfunc='sum')
```

```python
table = pd.pivot_table(data=df,index=['Gender', 'Religion'], aggfunc={'Yearly Tra':np.sum})
table
```

```python
table.plot(kind='bar')
```

```python
table = pd.pivot_table(data=df,index=['Gender'], columns=["Religion"],values=['Yearly Tra'],aggfunc={'Yearly Tra':np.sum})
table
```

```python
table.plot(kind='bar')
```

```python
table = pd.pivot_table(data=df,index=['Gender'], columns=["Occupation"],values=['Yearly Tra'],aggfunc={'Yearly Tra':np.sum})
table
```
